```python
def booking(seats):
    n=int(input("Enter the No of Seats:"))
    l=[]
    for i in range(0,n):
        seat_no=input("Enter the Seat No:")
        if seats[seat_no]!=True:
            print(f"Your {seat_no} Selected seat is already booked")
            print("re-enter")
            return booking()
        else:
            l+=[seat_no]
    d={}
    for i in l:
        print("Your seat no:",i)
        name=input("Enter the name:")
        age=input("Enter the age")
        gender=input("Enter the gender")
        if int(i[-1])%2==0 and seats[i[0]+str(int[-1])-1]!=True:
            if seats[i[0]+str(int(i[-1])-1)][1]['gender'].upper()==gender.upper():
                d[i]=(False,{'name':name,'age':age,'gender':gender})
            else:
                print(i,f"seat should be {seats[i[0]+str(int(i[-1])+1)][1]['gender']}")
                return booking() 
        elif int(i[-1])%2==0 and seats[i[0]+str(int[-1])-1]!=True:
            if seats[i[0]+str(int(i[-1])-1)][1]['gender'].upper()==gender.upper():
                d[i]=(False,{'name':name,'age':age,'gender':gender})
            else:
                print(i,f"seat should be {seats[i[0]+str(int(i[-1])+1)][1]['gender']}")
                return booking() 
        else:
            d[i]=(False,{'name':name,'age':age,'gender':gender})
    return d    
```
